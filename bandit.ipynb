```python
import numpy as np
from multi_armed_bandit import k_bandit_epsilon_greedy, plot_average_reward
```

### Run bandit and plot outcomes

As expected, average reward increases as the number of iterations increase

```python
# Define reward means and variances so that the same values are used in each bandit
num_actions = 10
reward_means = np.random.normal(0, 1, num_actions)
reward_variances = np.repeat(1, num_actions)
```

```python
# increase number of iterations for reward
average_reward_dict = {}
for num_iterations in range(10, 1000, 10):
    avg_reward_list = []
    for i in range(200): # Run 200 bandits
        bandit = k_bandit_epsilon_greedy(k=num_actions, 
                                         epsilon=0.1, 
                                         iters=num_iterations, 
                                         reward_means=reward_means, 
                                         reward_variances=reward_variances)
        total_reward = bandit.run_bandit()
        avg_reward = total_reward/num_iterations
        avg_reward_list.append(avg_reward)
    average_reward_dict[num_iterations] = sum(avg_reward_list)/len(avg_reward_list)
```

```python
bandit.plot_reward_means()
```

```python
plot_average_reward(average_reward_dict, 0.1)
```

### Explore how curve changes with several modification.

1. Modify epsilon value to 1% (compared to 10% above). 


```python
# increase number of iterations for reward
average_reward_dict_2 = {}
for num_iterations in range(10, 1000, 10):
    avg_reward_list = []
    for i in range(200): # Run 200 bandits
        bandit = k_bandit_epsilon_greedy(k=num_actions, 
                                         epsilon=0.01, 
                                         iters=num_iterations, 
                                         reward_means=reward_means, 
                                         reward_variances=reward_variances)
        total_reward = bandit.run_bandit()
        avg_reward = total_reward/num_iterations
        avg_reward_list.append(avg_reward)
    average_reward_dict_2[num_iterations] = sum(avg_reward_list)/len(avg_reward_list)
```

```python
plot_average_reward(average_reward_dict_2, 0.01)
```

2. Start with higher initial values for reward means. (This should result in more exploration.)

    - This exploration increases average reward!

```python
average_reward_dict_3 = {}
for num_iterations in range(10, 1000, 10):
    avg_reward_list = []
    for i in range(200): # Run 200 bandits
        bandit = k_bandit_epsilon_greedy(k=num_actions, 
                                         epsilon=0.1, 
                                         iters=num_iterations, 
                                         initial_reward=5.0,
                                         reward_means=reward_means, 
                                         reward_variances=reward_variances)
        total_reward = bandit.run_bandit()
        avg_reward = total_reward/num_iterations
        avg_reward_list.append(avg_reward)
    average_reward_dict_3[num_iterations] = sum(avg_reward_list)/len(avg_reward_list)
```

```python
plot_average_reward(average_reward_dict_3, 0.1)
```
